```python
import pandas as pd
import numpy as np
```

```python
# Loading the Data
Dashen_Data = pd.read_csv('../data/raw/Dashen_Bank_reviews_20250608_154622.csv')
Abyssinia_Data = pd.read_csv('../data/raw/Bank_of_Abyssinia_reviews_20250608_154623.csv')
Commercial_Data = pd.read_csv('../data/raw/Commercial_Bank_of_Ethiopia_reviews_20250608_154621.csv')
```

```python
# Data Overview and Description
def data_overview(data):
    print("Data Overview:")
    print(data.head())
    print("\nSize of the Dataset:")
    print(data.shape)
    print("\nData Description:")
    print(data.describe())
    print("\nData Info:")
    print(data.info())
    print("\nMissing Values:")
    print(data.isnull().sum())
    print("\nUnique Values in Each Column:")
    for column in data.columns:
        print(f"{column}: {data[column].nunique()}")
data_overview(Dashen_Data)
```

```python
# Data Overview and Description
def data_overview(data):
    print("Data Overview:")
    print(data.head())
    print("\nSize of the Dataset:")
    print(data.shape)
    print("\nData Description:")
    print(data.describe())
    print("\nData Info:")
    print(data.info())
    print("\nMissing Values:")
    print(data.isnull().sum())
    print("\nUnique Values in Each Column:")
    for column in data.columns:
        print(f"{column}: {data[column].nunique()}")
data_overview(Commercial_Data)
```

```python
#column overview
def column_overview(data):
    print("\nColumn Overview:")
    for column in data.columns:
        print(f"Column: {column}")
        print(f"Data Type: {data[column].dtype}")
        print(f"Number of Unique Values: {data[column].nunique()}")
        print(f"Sample Values: {data[column].dropna().unique()[:5]}")
        print("-" * 40)
column_overview(Dashen_Data)
```

```python
# Function to check null values and print with dataset name
def check_null_values(data, name):
    print(f"{name} Null Values:\n{data.isnull().sum()}\n")

# Call it for each bank
check_null_values(Dashen_Data, "Dashen")
check_null_values(Commercial_Data, "Commercial")
check_null_values(Abyssinia_Data, "Abyssinia")
```

```python
# NOrmalizing the data of Date in the dataset
def normalize_date(data):
    data['date'] = pd.to_datetime(data['date'], errors='coerce')
    data['Year'] = data['date'].dt.year
    data['Month'] = data['date'].dt.month
    data['Day'] = data['date'].dt.day
    return data
Dashen_Data = normalize_date(Dashen_Data)
Abyssinia_Data = normalize_date(Abyssinia_Data)
Commercial_Data = normalize_date(Commercial_Data)
```

```python
print("Dashin Data:\n",Dashen_Data.head())
print("Commercial_Data Data:\n",Commercial_Data.head())
print("Abyssinia_Data Data:\n",Abyssinia_Data.head())
```

```python
# Save the cleaned data to CSV files
def save_to_csv(data, filename):
    data.to_csv(filename, index=False)
    print(f"Data saved to {filename}")
# Saving the cleaned data to CSV files
save_to_csv(Dashen_Data, '../data/processed/Dashen_Bank_reviews_cleaned.csv')
save_to_csv(Abyssinia_Data, '../data/processed/Bank_of_Abyssinia_reviews_cleaned.csv')
save_to_csv(Commercial_Data, '../data/processed/Commercial_Bank_of_Ethiopia_reviews_cleaned.csv')
```
